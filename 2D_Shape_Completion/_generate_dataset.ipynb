Only needs to be run once when there is a folder /data_new containing segmaps and random samplings should be created from them. Samplings are saved in /data/samples.

```python
import os
import random
from dataset.preprocessing import pil_to_binary, get_surface_pixels, sample_pixels
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Directory paths
input_dir = './segmaps'
train_dir = './data/train'
sample_dir = "./data/train/samples"
test_dir = "./data/test"
test_sample_dir = "./data/test/samples"

# Ensure the output directory exists
os.makedirs(train_dir, exist_ok=True)
os.makedirs(sample_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(test_sample_dir, exist_ok=True)

# Randomly sample 10000 images from the input directory
file_list = random.sample(os.listdir(input_dir), 10000)
for i, filename in enumerate(file_list):
    if random.random() < 0.8:
        output_dir = train_dir
        save_dir = sample_dir
    else:
        output_dir = test_dir
        save_dir = test_dir
    # log every 100 files
    if i % 100 == 0:
        print(f"Processing file {i}")
    if filename.endswith('.png'):
        image_number = filename.split('.')[0]
        # Create subfolders for train and test
        os.makedirs(os.path.join(save_dir, image_number), exist_ok=True)
        
        # Load the image
        image_path = os.path.join(input_dir, filename)
        image = Image.open(image_path)
        binary_image = pil_to_binary(image)

        surface_pixels = get_surface_pixels(binary_image)

        output_path = os.path.join(output_dir, filename)
        plt.imsave(output_path, binary_image, cmap='gray')
        
        for i in range(100):
            #random number between 1 and 15
            rnd = random.randint(1,15)
            sampled_points = sample_pixels(surface_pixels, rnd)
                
            #save image to appropriate subfolder
            sample_path = os.path.join(save_dir, image_number, f'{i}.png')
            plt.imsave(sample_path, sampled_points, cmap='gray')

print("Sampling completed and saved to", sample_dir)
```
